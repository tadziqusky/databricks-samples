```
import folium as folium
import math as math
import pandas as pd
```

```
# generating sample data
data = {
  'Id': [1, 2, 3, 4, 5],
  'Latitude': [40.7518471, 40.7712888, 40.6894449, 40.6738445, 40.6566844],
  'Longitude': [-73.787862, -73.9466928, -73.9551212, 73.9250801, -73.9298964]
}

pdf = pd.DataFrame(data)
df = spark.createDataFrame(pdf)

# normally you would query the data from a source
result_pdf = df.select("*").toPandas()
display(df)
```

```
map_position = result_pdf[["Latitude", "Longitude"]].iloc[1,:]

map = folium.Map(location=[map_position["Latitude"], map_position["Longitude"]], zoom_start=11)

result_pdf.apply(lambda row: 
                folium.CircleMarker(
                  location=[row["Latitude"], row["Longitude"]], 
                  radius=5, 
                  popup=f'{row["Id"]}'
                ).add_to(map), axis=1)
 
display(map)
```
